```python
import cv2
import numpy as np
import os
import csv


router = './data-release/train/' 
classifire = './data-release/traindata/' 



with open('./data-release/train_labels.csv') as csvfile:

   
    if not os.path.exists(classifire):
        os.mkdir(classifire)

   
    rows = csv.reader(csvfile)

    for row in rows:
        
        try:
            files = '{}/{}'.format(router,row[0])
            files2  = classifire + '{}'.format(row[1]) 
            
            
            if row[0] == 'id':
                continue
            
            if not os.path.exists(files2):
                os.mkdir(files2)
            
            img1 = cv2.imread(files+'_c.png')
            img2 = cv2.imread(files+'_v.png')
            

            
            image = np.concatenate((img1, img2))  
            

            
            cv2.imwrite('{}/{}.png'.format(files2,row[0]), image)

        except:
            continue



router = './data-release/test/'
classifire = './data-release/testdata/' 



with open('./data-release/submission_format.csv') as csvfile:

  
    if not os.path.exists(classifire):
        os.mkdir(classifire)


    rows = csv.reader(csvfile)

    for row in rows:
        try:
            files = '{}/{}'.format(router,row[0]) 

           
            img1 = cv2.imread(files+'_c.png')
            img2 = cv2.imread(files+'_v.png')
          

           
            image = np.concatenate((img1, img2)) 
           
            cv2.imwrite('{}/{}.png'.format(classifire,row[0]), image)

        except:
            continue
```
